```
#code for investment calculator
#T-Level Sample Assessment Material (Task 2)
#Non working code

def opening ():
    print('#####################################')
    print('Welcome to the investment quote system')
    print('')
    print('Please enter your name')
    opening.name = input()
    
    print('Please enter your Address')
    opening.address = input()
    
    print ('Please enter your telephone number')
    opening.phone = input()
    
    print ('How much would you like to invest per month (£)?')
    while True:
      opening.investSum = input()
      if (opening.investSum.isnumeric()):
        opening.investSum = int(opening.investSum)
        break

def options ():
    print('#####################################')
    print('There are two types of investment available:')
    print('Option 1 - Savings plan')
    print('Option 2 - Managed stock investment')
    print('Please select an option (press 1 or 2 followed by enter)')
    print('#####################################')
    
    option = input()
      
    while option != '1' and option != '2':
        print('Please select an option (press 1 or 2 followed by enter)')
        
        option = input()
        
    if option == '1':
        savingsMain()
    else:
        stocksMain()
        

def savingsMain():
    savingsMain.monthlyInvest = opening.investSum
    savingsMain.yearlyInvest = savingsMain.monthlyInvest * 12
    
    while savingsMain.yearlyInvest > 20000:
        print('The initial monthly amount is too high for this type of plan' )
        print ('How much would you like to invest per month (£)?')
        savingsMain.monthlyInvest = float(input())
        savingsMain.yearlyInvest = savingsMain.monthlyInvest * 12
    
    savingsPrint()
    
def savingsMin():
    predictReturns = 0.012
    yearlyFees = 0.0025 * 12
    savingsMin.total = savingsMain.yearlyInvest
    print('#####################################')   
    print('Forecasted performance of this plan at the lowest rate of return:')
    runningFees = 0
    runningEarnings = 0
    for i in range(10):
        earnings = savingsMin.total * predictReturns
        runningEarnings = runningEarnings + earnings
        fees = savingsMin.total * yearlyFees
        runningFees = runningFees + fees
        savingsMin.total = savingsMin.total + earnings - fees
        
        if i == 0 or i == 4 or i == 9:
            print('At the end of year', str(i+1))
            print('Your investment will be worth:')
            print('£', round(savingsMin.total,2))
            print('')
            print('Total fees paid in this period: £', round(runningFees, 2))
            print('')
            print('Total profit in this period: £', round(savingsMin.total - savingsMain.yearlyInvest, 2))
            print('')
    print('#####################################')
    print('')

def savingsMax():
    predictReturns = 0.024
    yearlyFees = 0.0025 * 12
    savingsMax.total = savingsMain.yearlyInvest
    print('#####################################')
    print('Forecasted performance of this plan at the highest rate of return::')
    runningFees = 0
    runningEarnings = 0
    for i in range(10):
        earnings = savingsMax.total * predictReturns
        runningEarnings = runningEarnings + earnings
        fees = savingsMax.total * yearlyFees
        runningFees = runningFees + fees
        savingsMax.total = savingsMax.total + earnings - fees
         
        if i == 0 or i == 4 or i == 9:
            print('At the end of year', str(i+1))
            print('Your investment will be worth:')
            print('£', round(savingsMax.total,2))
            print('')
            print('Total fees paid in this period: £', round(runningFees, 2))
            print('')
            print('Total profit in this period: £', round(savingsMax.total - savingsMain.yearlyInvest, 2))
            print('')
    print('#####################################')  
    print('')
     
def stocksMain():
    stocksMain.monthlyInvest = opening.investSum
    stocksMain.yearlyInvest = stocksMain.monthlyInvest * 12
    
    stocksPrint()
    
def stocksMin():
    predictReturns = 0.04
    yearlyFees = 0.013 * 12
    stocksMin.total = stocksMain.yearlyInvest
    print('#####################################')   
    print('Forecasted performance of this plan at the lowest rate of return:')
    
    runningFees = 0
    runningEarnings = 0
    runningTax = 0
    for i in range(10):
        earnings = stocksMin.total * predictReturns
        runningEarnings = runningEarnings + earnings
        fees = stocksMin.total * yearlyFees
        runningFees = runningFees + fees
        
        if stocksMin.total > 40000:
            taxRate = 0.2
        elif stocksMin.total > 12000:
            taxRate = 0.1
        else:
            taxRate = 0
        
        taxPayable = stocksMin.total * taxRate
        runningTax = runningTax + taxPayable
        stocksMin.total = stocksMin.total + earnings - fees

        if i == 0 or i == 4 or i == 9:
            print('At the end of year', str(i+1))
            print('Your investment will be worth:')
            print('£', round(stocksMin.total,2))
            print('')
            print('Total fees paid from start: £', round(runningFees,2) )
            print('')
            print('Total profit from start: £', round(stocksMin.total - stocksMain.yearlyInvest, 2))
            print('')
            print('Total tax due from start: £', round(runningTax, 2))
            print('')
    print('#####################################')
    print('') 

def stocksMax():
    predictReturns = 0.23
    yearlyFees = 0.013 * 12
    stocksMax.total = stocksMain.yearlyInvest
    print('#####################################')   
    print('Forecasted performance of this plan at the higher rate of return:')
    
    runningFees = 0
    runningEarnings = 0
    runningTax = 0
    for i in range(10):
        earnings = stocksMax.total * predictReturns
        runningEarnings = runningEarnings + earnings
        fees = stocksMax.total * yearlyFees
        runningFees = runningFees + fees
        
        if stocksMax.total > 40000:
            taxRate = 0.2
        elif stocksMax.total > 12000:
            taxRate = 0.1
        else:
            taxRate = 0
        
        taxPayable = stocksMax.total * taxRate
        runningTax = runningTax + taxPayable
        stocksMax.total = stocksMax.total + earnings - fees - taxPayable

        if i == 0 or i == 4 or i == 9:
            print('At the end of year', str(i+1))
            print('Your investment will be worth:')
            print('£', round(stocksMax.total, 2))
            print('')
            print('Total fees paid from start: £', round(runningFees, 2))
            print('')
            print('Total profit from start: £', round(stocksMax.total - stocksMain.yearlyInvest, 2))
            print('')
            print('Total tax due from start: £', round(runningTax, 2))
            print('')
    print('#####################################')
    print('') 

def savingsPrint ():
    print('--------------------------------------------------------')
    print('Personal Investment Quote for:')
    print('Name: ', opening.name)
    print('')
    print('Address: ', opening.address)
    print('')
    print('Telephone Number: ', opening.phone)
    print('--------------------------------------------------------')
    print('')
    print('You selected a savings plan')
    savingsMin()
    savingsMax()

def stocksPrint ():
    print('--------------------------------------------------------')
    print('Personal Investment Quote for:')
    print('Name: ', opening.name)
    print('')
    print('Address: ', opening.address)
    print('')
    print('Telephone Number: ', opening.phone)
    print('--------------------------------------------------------')
    print('')
    print('You chose a Managed Stock Investment plan')
    stocksMin()
    stocksMax()
```
